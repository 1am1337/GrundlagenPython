```python
import numpy as np
import matplotlib.pyplot as plt
import librosa as lr
from IPython.display import Audio
```

```python
a=np.array([[0,0,1,0,0],
          [0,1,0,1,0],
          [1,0,0,0,1],
          [0,1,0,1,0],
          [0,0,1,0,0]])
b=np.linspace(0,1)
```

```python
sr = 44100 #samplerate in hz
length = 2 # in sec
freq1 = 2000 # in hz
freq2 = 2210 # in hz
freq3 = 1890 # in hz
t = np.linspace(0, length, num = sr * length)

y = np.sin(2*np.pi*t*freq1)
y = y + np.sin(2*np.pi*t*freq2)
y = y + np.sin(2*np.pi*t*freq3)
#normalisierung
y = y / np.max(np.abs(y))
plt.plot(t,y)
plt.xlim(0,0.5)
plt.show()
display(Audio(y, rate = sr))
```

```python
stft = lr.stft(y=y)
plt.imshow(np.abs(stft), "Greys",aspect="auto", origin="lower")
plt.ylabel("placeholder")
plt.xlabel('time')
plt.ylim(75, 125)
plt.show()
```
