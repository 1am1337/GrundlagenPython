```python
x = np.arange(100) /100
plt.plot(x*2*np.pi, np.sin(x*2*np.pi), ".")
plt.xlabel("radianten")
plt.ylabel("amplitude")
plt.show()
```

```python
sr = np.arange(48000) /48000
freq = 50 #hz
plt.plot(sr, np.sin(freq*x*2*np.pi), "-")
plt.xlabel("zeit")
plt.ylabel("amplitude")
plt.show()
```

```python
sr = np.arange(48000) /48000
freq = np.arange(11)
for freq in freq:
    plt.title(freq)
    plt.plot(sr, np.sin(freq*sr*2*np.pi), "-")
    plt.show()
```
